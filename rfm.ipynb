```python
from sklearn.cluster import KMeans
```

```python
x = vitrina_per.drop(columns = ['OrderDate', 'Phone_new'])
```

```python
ess = []
for k in range(2, 12):
    kmeans = KMeans(n_clusters=k, random_state=0, max_iter=100)
    kmeans.fit(x)
    ess.append(kmeans.inertia_)

plt.rcParams["figure.figsize"] = (10,5)
plt.plot(range(2, 12), ess)
plt.xticks(range(2, 12))
plt.show()
```

```python
k = 5
kmeans = KMeans(n_clusters=k, random_state=0)
y = kmeans.fit_predict(x)
x
```

```python
df_kmeans = vitrina_per
df_kmeans['cluster'] = y
df_kmeans
```

```python
# Each cluster's fraction of the dataset
cluster_counts = np.array(df_kmeans['cluster'].value_counts().sort_index())
cluster_frac = cluster_counts/np.sum(cluster_counts)
cluster_frac
```

```python
df_kmeans.groupby(by = 'cluster').median().round(3)
```

```python
rfm_df = vitrina[['Phone_new', 'OrderDate', 'Number of bills', 'Revenue']].dropna()
```

```python
rfm_df.describe().T
```

```python
rfm_df['R_rank'] = rfm_df['OrderDate'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Number of bills'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Revenue'].rank(ascending=True)

rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['M_rank']/rfm_df['M_rank'].max())*100

rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

rfm_df
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28 * \
   rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
data_seg_RFM = rfm_df.round(2)
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] >
                                      4.5, "Top Customers",
                                      (np.where(
                                        rfm_df['RFM_Score'] > 4,
                                        "High value Customer",
                                        (np.where(
  rfm_df['RFM_Score'] > 3,
                             "Medium Value Customer",
                             np.where(rfm_df['RFM_Score'] > 1.6,
                            'Low Value Customers', 'Lost Customers'))))))
rfm_df[['RFM_Score', 'Customer_segment']].head(20)
```

```python
plt.pie(rfm_df.Customer_segment.value_counts(),
        labels=rfm_df.Customer_segment.value_counts().index,
        autopct='%.0f%%')
plt.show()
```

```python
for i in ['Top Customers', 'High value Customer', 'Medium Value Customer', 'Low Value Customers', 'Lost Customers']:
  print(i)
  print(vitrina[rfm_df['Customer_segment'] == i].describe().T)
```
